```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data)
df
```

```python
boston.feature_names
```

```python
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
df
```

```python
boston.target.shape
```

```python
df['Price'] = boston.target
```

```python
# # Input Data
X = boston.data

# # Output Data
y = boston.target
```

```python
# Spliting the Data in two
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3,random_state = 45)
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("***********Traing Complete************")
```

**Observation** - *It Slipts the data into two for testing and training of the model.*

```python
# predicting the test set results
y_pred = regressor.predict(X_test)
y_pred
```

```python
sns.regplot(y_test, y_pred)
plt.scatter(y_test, y_pred)
plt.xlabel("Price (in $1000's)")
plt.ylabel("Predicted Value")
plt.title("True Value vs Predicted Value")
plt.show()
```

**Observation** - *It shows the deviation of values from the regression line*

```python
# Calculating the accuracy of the model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:' ,metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("R-2:" ,metrics.r2_score(y_test,y_pred))
```

```python
regressor.score(X_test,y_test)
```

**Observation** - *The Accuracy of the model is 74%.*
