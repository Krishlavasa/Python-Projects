```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
# load dataset
df = pd.read_csv("D:\COLLEGE DATA\SEM III\BDS371L - Data Analytics Lab\GRASIM.csv")
df
```

**Observation** - *Importing the dataset from csv file in the dataset.*

```python
df.describe()
```

**Observation** - *Gives the general description about the dataset such as count, mean, std, min, max and quatiles.*

```python
df.info()
```

**Observation** - *Information about the columns and the number of counts*

```python
plt.style.use('fivethirtyeight')
df['Date'] = pd.to_datetime(df['Date'])
stockprice_one_year = df.loc[(df['Date'] >= '2007-01-04') & (df['Date'] <= '2008-03-31')]
stockprice_one_year
```

```python
plt.rcParams['figure.figsize'] = (6,6)
```

```python
df['Turnover'].plot(linewidth = 2 ,color = "red")
stockprice_one_year['Turnover'].plot(linewidth = 2 ,color = 'Black')
```

**Observation** - *The black part on the x-axis between 1500 and 2500 which denotes as the financial year 2007-2008.*

```python
plt.plot(stockprice_one_year['High'],linewidth = 2)
plt.plot(stockprice_one_year['Low'],linewidth = 2)
```

**Observation** - *The stockpriceoneyear is high till 4000 and low till 3800.*

```python
plt.plot(df['High'],linewidth = 2)
plt.plot(df['Low'],linewidth = 2)
```

**Observation** - *The stock price of df is high is above 5000 and low is also approx to 5000.*

```python
plt.figure(figsize=(9,9))
sns.heatmap(stockprice_one_year.corr(),annot= True)
plt.style.use('fivethirtyeight')
plt.title('HeatMap')
```

**Obervation** - *Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship.*

```python
plt.figure(figsize=(6, 6))
x = stockprice_one_year['Date']
y = stockprice_one_year['Low']
plt.plot(x,y,linewidth = 2)
plt.style.use('fivethirtyeight')
plt.title("Stock Price - LOW")
```

**Observation** - *It shows the low priceso of the stock over one year in the dataset.*

```python
half1 = stockprice_one_year.loc[(df["Date"] > "2008-02-02")]
half1
```

```python
plt.fill_between(x=stockprice_one_year["Date"],y1= stockprice_one_year["High"],y2=stockprice_one_year["Low"],fc = "Grey")
plt.plot(stockprice_one_year["Date"],stockprice_one_year["High"],color= "Green",)
plt.plot(stockprice_one_year["Date"],stockprice_one_year["Low"],color= "Red",)
```

**Observation** - *fill_between() is used to fill area between two horizontal curves as the stockprice_one_year of both High and Low prices of the dataset*

```python
plt.plot(stockprice_one_year["Date"],stockprice_one_year["Close"],color= "Red")
plt.plot(half1["Date"],half1["Close"],color= "Green")
```

**Observation** - *stockprice_one_year of close is shown in red color*

```python
sns.boxplot(data=stockprice_one_year[["High","Low","Close","Open"]])
```

**Observation** - *Boxplot medium of High, Low, Close, Open between the range between 2500 - 3000*
