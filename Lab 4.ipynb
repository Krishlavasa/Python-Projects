```python
# Importing the Libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# load dataset
df = pd.read_csv("df1.csv")
df
```

**Observation** - *Importing the dataset from csv file form the last saved dataset*

```python
plt.bar(df['Duration(Hrs.)'],df['Fare'])
plt.xlabel("Duration(Hrs.)")
plt.ylabel("Fare(Rs.)")
plt.title("Relationship between Duration and Fare")
```

**Observation** - *Relationship between the Fares(Rs.) and Duration(Hrs.) because it's closely related to it as the fuel consumption is observed while duration of the journey*

```python
sns.scatterplot(x="Gender",y="Number_of_tickets",data=df,palette='pastel',legend='brief')
```

**Observation** -*Shows how the Genders are related with the Number_of_tickets in the dataset*

```python
sns.catplot(x='Gender',y='Fare',kind ='bar',data = df)
```

**Observation** - *Camparing the Fare with the Gender in the dataset*

```python
sns.catplot(x='Duration(Hrs.)',kind = 'count',data = df)
```

**Observation** - *Camparing the Duration(Hrs.) by the count in the dataset*

```python
d_df = df.where(df['Fare'] < 700).dropna()
d_df
```

**Observation** - *Dropping the Fares who are below 700 in the dataset*

```python
d_df
```

```python
sns.pairplot(df)
```

**Observation** - *There is no correlation in the dataset so there are no inferences*

```python
sns.pairplot(df,hue ='Class_type')
```

**Observation** - *There is no correlation in the dataset so there are no inferences*

```python
sns.violinplot(x ="Number_of_tickets", y ="Fare",data = df)
```

**Observation** - *Voilin Plot shows the median as 1500 in the dataset*

```python
sns.stripplot(x ="Number_of_tickets", y ="Fare",data=df)
plt.title('Strip Plot')
```

**Observation** - *There is a strip at each number of tickets in the dataset*

```python
sns.lineplot(x ="PNR", y ="Fare", data=df)
plt.title('Line Plot')
```

**Observation** - *Line plot the values of fares lies between 2000 and 1000.*

```python
sns.lineplot(x ="Number_of_tickets", y ="Fare",data= df)
plt.title('Line Plot')
```

**Observation** - *The average fare and number of tickets are somewhat between 1560 to 1460 in the dataset*

```python
sns.countplot(x ='Destination', data = df)
plt.title('Count Plot')
```

**Observation** - *The highest number of Destination is "Lucknow"  present in the dataset*

```python
sns.heatmap(df.corr())
plt.title('HeatMap')
```

**Observation** - *There is no correlation in the dataset so there are no inferences*

```python
sns.PairGrid(df,hue= 'Fare')
plt.title('Pair Grid')
```

**Observation** - *There is no corrlation in the dataset so there are no inferences*

```python
df.corr()
```

```python
sns.jointplot(x ="Number_of_tickets", y ="Fare", data= df)
plt.title('Joint Plot')
```

**Observation** - *Gives the infrences between number of tickets and fare in the dataset*

```python
sns.jointplot(x =df["Number_of_tickets"], y =df["Fare"], hue= df['Gender'])
plt.title('Joint Plot')
```

**Observation** - *Jointplot relationship between fare and Number of tickets with refernce to Gender in the dataset*

```python
sns.boxplot(x=df['Duration(Hrs.)'])
plt.title('Box Plot')
```

**Observation** - *The mid point 20 inferenced from in the dataset*
