```python
# Importing the Libraries
import pandas as pd
from matplotlib import pyplot as plt
import random
from random import randint
df1 = pd.DataFrame(columns=['PNR','Number_of_tickets','Class_type','Source','Duration(Hrs.)','Destination','Gender','Fare'])
```

```python
# Creating Dictionary and declaring 
new_train= {}
cl = ['1AC_class','2AC_class','3AC_class','Sleeper_class','2-S_seating']
sor = ['Delhi','Pune','Mumbai','Kerela']
des = ['Lucknow','Kochi','Gurugram','Assam']
gen = ['Male','Female']

for i in range(0,2000):
    new_train['PNR'] = randint(2423522587,2423522888)
    new_train['Number_of_tickets'] = randint(1,4)
    new_train['Class_type'] = cl[randint(0,4)]
    new_train['Source'] = sor[randint(0,3)]
    new_train['Destination'] = des[randint(0,3)]
    new_train['Duration(Hrs.)'] = randint(10,30)
    new_train['Gender'] = gen[randint(0,1)]
    new_train['Fare'] = randint(500,2500)
    df1 = df1.append(new_train,ignore_index= True)
```

```python
#Sample 20% of the Dataset and Create a new dataframe
random.seed(100)
df2 = pd.DataFrame.sample(df1,400)
df2
```

**Obervation** - *Creating new dataframe and taking out 20% of the previous dataset by using sample() function*

```python
# Removing the Rows based on the Condition 
df2.drop(df2[df2['Duration(Hrs.)'] < 15].index,inplace =True)
df2
```

**Obervation**  - *Removing the rows from the Duration(Hrs.) column which are less than 15*

```python
# Replace some values of one column with 'NaN' or 'Null'
df2['Duration(Hrs.)'] = df2['Duration(Hrs.)'].replace([11],None)
df2['Duration(Hrs.)'] = df2['Duration(Hrs.)'].replace([24],None)
df2['Duration(Hrs.)'] = df2['Duration(Hrs.)'].replace([20],None)
df2['Duration(Hrs.)'] = df2['Duration(Hrs.)'].replace([15],None)
df2
```

**Obervation**  - *Replacing the values in Duration(Hrs.) column with the values NA values to create missing values in the dataset*

```python
# Find the difference between .fillna(), .interpolate() and .dropna()
# .fillna()
df2['Duration(Hrs.)'].fillna(29,inplace = True)
df2
```

```python
#Categorising the values on the basis of specific parameters and dividing the column entries into 70% , 20%, 10% of the specific column
def randTickets():
    rand = (randint(0,100))/100
    
    if 0<rand<0.7:
        return "A"
        
    if 0.7<rand<0.9:
        return "B"

    else:
        return "C"
```

```python
Holder = []
for x in range(1000):
    value = randTickets()
    Holder.append(value)
```

```python
Holder.count("A")
```

```python
Holder.count("B")
```

```python
Holder.count("C")
```
