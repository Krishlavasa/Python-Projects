### Import the Toy IRIS Dataset

```python
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['Species'])
```

**Observation** - *Successfully imported the dataset from sklearn library*

### Print the Head and Tail of the Dataset (first 5 and last 5 entries)

```python
df.head()
```

```python
df.tail()
```

**Observation** - *First five rows are printed using head and last five rows are printed using tail functions respectively*

### Create a New Column named "Average" and find the Average of existing four features and store the same

```python
df['Average'] = df.mean(numeric_only=True, axis=1)
df
```

**Observation** - *A new column named with average is added to the dataset containing the average of each row of the dataset*

### Create a New Column named "Sum" and find the Sum of existing four features and store the same

```python
df['Sum'] = df.sum(numeric_only=True, axis=1)
df
```

**Observation** - *A new column named with Sum is added to the dataset containing the sum of each row of the dataset*

### Check how many entires of "IRIS Setosa", "IRIS Virgnica" and "IRIS Versicolor" are present in the Dataset

```python
iris['target_names']
```

```python
df['Species'] = df['Species'].map(dict(enumerate(iris['target_names'])))  
df
#Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object.
```

```python
df["Species"].value_counts()
```

**Observation** - *The Species Column is remaped with their names and then counting their values*

### Find the Descriptive Statistics about all the Six Columns (existing four + two made-up features)

```python
df.describe()
```

**Observation** - *Descriptive Statistics shows the count, mean, standard deviation, min, max, 25%, 50%,75% of each column of the dataset which helps us to gain meaningful insights* <br>
*The mean and Standard Deviation of Sepal_Length column is higher than all columns*

### Find the entries whose Sum Values are more than a User Input

```python
n = int(input("Enter the number to check Sum: "))
df[df['Sum'] > n]
```

**Observation** - *Entries who's value is greater than input are showcased in the output*

### Find the entries whose Sepal_Length is greater than Petal_Width

```python
df[df["sepal length (cm)"] > df["petal width (cm)"]]
```

**Observation** - *Entries who's Sepal Length is greater than Petal Width are showcased in the output*
