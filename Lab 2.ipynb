```python
df['Survival_Status'] = df.Survival_Status.map({0:'Not Survived ', 1:'Survived'})
df.Survival_Status.value_counts()
```

**Observation** - *The Number of Survived and not Survived People are **545** and **342** Respectively.*

```python
freetravellers =df[df['Fare_Paid']==0]
print(freetravellers.head())
print("The Number of Passengers who travelled for FREE : ", len(freetravellers))
```

**Observation** - *The numbers of Passengers who travelled for free are 15* 

```python
passenger_income = []
for i in df['Prices_Inr']:
    if i > (mean(df['Prices_Inr'])):
        passenger_income.append("Above Average")
    else:
        passenger_income.append("Below Average")
df['Passenger_Income'] = passenger_income
df
```

```python
df['Passenger_Income'].value_counts()
```

**Observation** - *The Income of Passengers were categorised based on two factors Below and Above Average*

```python
table = pd.crosstab(df['Survival_Status'],df['Gender'])
print (table)
```

**Observation** - *The number of Males and  females who survived and not survived*

```python
class1 = df[(df['Pclass'] == 1) & (df['Survival_Status'] == 'Survived') & (df['Gender'] == 'male')]
len(class1)
```

**Observation** - *The Male Passengers who survived in Class1 are 45*

```python
class2 = df[(df['Pclass'] == 2) & (df['Survival_Status'] == 'Survived') & (df['Gender'] == 'male')]
len(class2)
```

**Observation** - *The Male Passengers who survived in Class2 are 17*

```python
class3 = df[(df['Pclass'] == 3) & (df['Survival_Status'] == 'Survived') & (df['Gender'] == 'male')]
len(class3)
```

**Observation** - *The Male Passengers who survived in Class3 are 47*

```python
age_category = []
for i in df['Age']:
    if i < 18:
        age_category.append("Child")
    elif i > 18 and i < 60:
        age_category.append("Adult")
    else:
        age_category.append("Senior Citizen")
df['Age_category'] = age_category
df
```

```python
df['Age_category'].value_counts()
```

**Observation** - *The People are categorised based on Adult,Child and Senior Citizen respectively.*

```python
df = df.drop('Fare_Paid' ,axis='columns')
df
```

**Observation** - *Droped the Fare_Paid column from the dataset*

```python
pclass_map = {1: 'Upper_class', 2: 'Middle_class', 3: 'Lower_class'}
df['Pclass'] = df['Pclass'].map(pclass_map)
table = pd.crosstab(df['Survival_Status'],df['Pclass'])
print (table)
```

**Observation** - *The number of Lower_Class ,Middle_class & Upper_class Survived and Not Survived , i.e. ,categorisation based upon classes*

```python
#The iterrows() method generates an iterator object of the DataFrame, allowing us to iterate each row in the DataFrame.
married = []
for index, row in df.iterrows():
    if row["Name"] == ("Mrs." ):
        married.append("Married")
    elif row["Name"] == ("Mr."):
        married.append("Married")
    else :
        married.append("Unmarried")
df["Married_Status"] = married
df
```

**Observation** - *The number of Married and unmarried Peoples*
