```python
import pandas as pd
import seaborn as sns
import pandas_profiling as pp
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
# load dataset
df = pd.read_csv("D:\COLLEGE DATA\SEM III\BDS371L - Data Analytics Lab\TreasureHuntDataset.csv")
df
```

```python
df.fillna(value = 0,axis=1,inplace= True)
df
```

```python
df['How many team mates did you know before?'].unique()
df['How many team mates did you know before?'] = df['How many team mates did you know before?'].replace(to_replace=['-', 'volunteers','nan', 'None', 'one','One', 'One ', 'NONE', 'none', 'Zero', '------Null'],value='5')
df["How many team mates did you know before?"].unique()
```

```python
df['How many team mates spoke the same language as yours?'] = df['How many team mates spoke the same language as yours?'].replace(to_replace=['-', 'volunteers', 'few ','None', 'NONE', 'none', 'Zero', '------Null','One'],value='0')
df['How many team mates spoke the same language as yours?'] = df['How many team mates spoke the same language as yours?'].replace(to_replace=['Everyone spoke the same language','Everyone spoke in English ','Everyone','All of them','all of them','all 5','all','All','Everytime','All have communicated in English sometimes hindi', 'None (all English ('],value = '5')
df["How many team mates spoke the same language as yours?"].unique()
```
